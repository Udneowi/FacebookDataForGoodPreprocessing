**Filter out non-DK**

```python
def point_inside_polygon(xy, poly):
    """Determine if points x and y are inside poly.
    Source: http://www.ariel.com.au/a/python-point-int-poly.html
    """
    (x, y) = xy
    n = len(poly)
    inside = False

    p1x, p1y = poly[0]
    for i in range(n + 1):
        p2x, p2y = poly[i % n]
        if y > min(p1y, p2y):
            if y <= max(p1y, p2y):
                if x <= max(p1x, p2x):
                    if p1y != p2y:
                        xinters = (y - p1y) * (p2x - p1x) / (p2y - p1y) + p1x
                    if p1x == p2x or x <= xinters:
                        inside = not inside
        p1x, p1y = p2x, p2y

    return inside

geofence = [
    [56.053763, 12.643370],
    [55.874394, 12.639808],
    [55.555933, 12.723673],
    [55.156461, 12.803561],
    [55.420191, 15.272335],
    [54.794979, 15.412107],
    [54.672100, 12.567657],
    [54.506657, 11.970314],
    [54.704103, 10.370764],
    [54.803260, 9.909312],
    [54.832522, 9.420309],
    [54.925024, 8.589416],
    [55.594882, 7.998546],
    [57.147320, 7.796872],
    [57.989550, 11.155343],
    [56.154584, 12.479144]
]

fencemask = [
    point_inside_polygon(data.loc[i, ['lat', 'lon']].values, geofence)
    for i in data.index
]
```

```python
data = data.loc[fencemask]
```

**Get tilesizes**

```python
sorted_lats = np.sort(data['lat'].unique())
latdiffs = sorted_lats[1:] - sorted_lats[:-1]
dlat = min(latdiffs)

sorted_lons = np.sort(data['lon'].unique())
londiffs = sorted_lons[1:] - sorted_lons[:-1]
dlon = min(londiffs)
```

**Build geojson**

```python
def get_tile(coord, dlat, dlon):
    return [
        coord[::-1],
        [coord[0]-dlat, coord[1]][::-1],
        [coord[0]-dlat, coord[1]-dlon][::-1],
        [coord[0], coord[1]-dlon][::-1],
        coord[::-1]
    ]
```
